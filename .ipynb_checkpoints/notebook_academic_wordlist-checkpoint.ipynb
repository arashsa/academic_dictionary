```python
from lxml import etree
from collections import Counter
import re

tree = etree.parse('noecon01.xml')
notags = etree.tostring(tree, encoding='utf-8', method='text')

notags = re.sub(ur'[^a-zA-Z0-9]',' ', notags, re.UNICODE)

word_list = []
for w in notags.split():
    word_list.append(w)

freq_dist = Counter(word_list)
print freq_dist
```
